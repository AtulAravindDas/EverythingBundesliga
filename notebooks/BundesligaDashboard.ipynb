```python
import requests
import pandas as pd

API_TOKEN='Insert your API token here'
API_URL='https://api.football-data.org/v4/competitions/BL1/standings'

params = {'season': '2024'}
response=requests.get(API_URL, headers={'X-Auth-Token': API_TOKEN},params=params)
bundesliga_standings_dict=response.json()
bundesliga_standings_df=pd.DataFrame(bundesliga_standings_dict['standings'][0]['table'])
bundesliga_standings_df
```

```python
bundesliga_standings_df['team']=bundesliga_standings_df['team'].apply(lambda x:x['name'])
bundesliga_standings_df
```

Clusters defining the teams that have qualified for European Football, Are midtable, or in relegation

```python
bundesliga_name_points=bundesliga_standings_df[['team','points']]
bundesliga_name_points
```

```python
def generate_cluster(index):
  if index<6:
    return "European Qualified"
  elif index>=6 and index<15:
    return "Midtable"
  elif index==15:
    return "Relegation Playoff"
  else:
    return "Relegation"

cluster=[]
for index in bundesliga_name_points.index:
  cluster.append(generate_cluster(index))
bundesliga_name_points['cluster']=cluster
bundesliga_name_points
```

```python
color_map={'European Qualified':'green','Midtable':'blue','Relegation Playoff':'orange','Relegation':'red'}
```

```python
import plotly.express as px

# Reset index so we can use it as x-axis
df = bundesliga_name_points.reset_index().rename(columns={'index': 'Team Index'})

# Create figure
fig = px.scatter(
    df,
    x='team',  # Changed from 'Team Index' to 'team'
    y='points',
    color='cluster',
    color_discrete_map=color_map,
    hover_data={
        'team': True,
        'points': True,
        'cluster': True,
        'Team Index': False
    },
    title="Bundesliga 2025 - Points by Team Cluster"
)

# Update layout with rotated team names
fig.update_layout(
    xaxis_title="Teams",
    yaxis_title="Points",
    legend_title="Cluster",
    template="plotly_white",
    xaxis={'categoryorder': 'array', 'categoryarray': df['team'].tolist()},  # Maintain table order
    xaxis_tickangle=-45  # Rotate team names 45 degrees
)

fig.show()
```

```python
bundesliga_standings_df = bundesliga_standings_df.merge(
    bundesliga_name_points[['team', 'cluster']],
    on='team',
    how='left'
)
```

```python
import folium

color_map = {
    'European Qualified': 'green',
    'Midtable': 'blue',
    'Relegation Playoff': 'orange',
    'Relegation': 'red'
}


germany_center = [51.1657, 10.4515]
map_bundesliga = folium.Map(location=germany_center, zoom_start=6)


for _, row in bundesliga_standings_df.iterrows():
    if pd.notnull(row['Latitude']) and pd.notnull(row['Longitude']):
        popup_text = f"""<b>Name:</b> {row['team']}<br>
                         <b>City:</b> {row['City']}<br>
                         <b>Points:</b> {row['points']}<br>
                         <b>Cluster:</b> {row['cluster']}"""

        cluster_color = color_map.get(row['cluster'], 'gray')  # Default to gray if cluster missing

        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=folium.Popup(popup_text, max_width=300),
            tooltip=row['team'],
            icon=folium.Icon(color=cluster_color, icon="info-sign")
        ).add_to(map_bundesliga)

map_bundesliga
```
